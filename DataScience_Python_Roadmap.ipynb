## Veri İnceleme Kavramları

*  Merkezi Eğilim Ölçüleri (Measures of Central Tendency) : bir veri setinin merkezini ya da tipik değerini ifade eder.   
     * Ortalama (Mean) : Tüm değerleri kapsar ve aykırı değerlerden etkilenir.
     * Medyan (Median) : Veriler sıralandığında ortada kalan değerdir ve aykırı değerlerden etkilenmez.
     * Mod (Mode) : En sık tekrar eden değerdir ve kategorik verilerde merkez ölçüsü olarak kabul edilir.  
  

* Dağılım Ölçüleri (Measures of Dispersion) : Verilerin merkezden ne kadar uzaklaştığını ve ne kadar yaygın olduklarını ölçer.
     * Standart Sapma : Verilerin ortalamadan ne kadar saptığını ölçer. Küçük std verilerin ortalamaya ve birbirine yakın olduğunu ifade eder
     * Varyans : Standart sapmanın karesidir, verilerin ortalamadan ne kadar farklılaştığını ölçer.
     * Aralık : En büyük değer (max) ile en küçük değer (min) arasındaki farktır ve değişim aralığını ifade etmektedir.
     * Kartiller (Interquartile Range - IQR) : Veri setini 4 parçaya bölerek yüzdelik dilimler ile dağılımını inceler. 
     * Çarpıklık (Skewness) : Verinin asimetrik olup olmadığını gösterir. Yani dağılımın sağa veya sola doğru eğilmiş olup olmadığını belirler.  
     * Basıklık (Kurtosis) : Verinin zirvesinin keskinliği veya düzlüğü hakkında bilgi verir.

* Kavramların Birbirleriyle İlişkileri
     * Ortalama, medyan, mod : Normal dağılımda bu ölçüler eşittir  
     fakat veriler çarpık (skewed) olduğunda, aykırı değerler ortalamayı etkiler;
     medyan ve mod ise bundan çok etkilenmez.  

     *  Standart sapma ve Ortalama: Küçük bir standart sapma verilerin ortalamaya yakın olduğunu, büyük standart sapma ise verilerin ortalamadan ne kadar uzak olduğunu işaret etmektedir.  
     
     * Dağılım Çarpıklığı (Skewness) :   
     Veriler sağa çarpıksa (pozitif çarpıklık) yani ortalama > medyan  
     Veriler sola çarpıksa (negatif çarpıklık) yani ortalama < medyan

     * Dağılımın Sivriliği (Kurtosis) :   
     Kurtosis = 3: Normal Dağılım (Mesokurtik) — Normal dağılıma yakın bir şekil.  
     Kurtosis > 3: Sivri (Keskin) Dağılım (Leptokurtik) — Zirve keskin, uç değerler daha fazla.  
     Kurtosis < 3: Basık Dağılım (Platykurtik) — Zirve düz, uç değerler daha az.  

     * IQR ve Aykırı Değerler: IQR, veri setindeki aykırı değerleri belirlemek için kullanılır.

Eksik Değerlerin İmputasyonu (Doldurulması) : Eksik değerleri tahmin etmek ve veri kaybını önlemek için çeşitli imputasyon yöntemleri kullanılır.

 * Basit İmputasyon : Ortalama, mod ve medyan kullanılarak doldurma işlemidir.

Doldurma yöntemini seçerken aykırı değerler vb. konulara dikkat ederek doldurma işlemi yapınız.

Eksik Veri Stratejisinin Seçimi  
Her proje için eksik veri stratejisi farklı olabilir. İşte bazı kriterler:  

* Eksik veri oranı düşükse (örneğin <%5), ortalama veya medyan imputasyonu kullanılabilir.  
* Eksik veri oranı yüksekse (örneğin >%30), sütunların tamamen kaldırılması veya daha gelişmiş yöntemler (KNN, regresyon vb) uygulanabilir.  
* Eksiklik analiz hedefini etkiliyorsa, eksik verilerin analiz sonuçlarına nasıl yansıdığı değerlendirilmelidir.  

```python
# Basit İmputasyon

# Ortalama ile doldurmak
df["Age"].fillna(df["Age"].mean(), inplace= True)


# Medyan ile doldurmak
df["Age"].fillna(df["Age"].median(), inplace=True)


# Mod ile doldurmak (Kategorik Veriler)
# Mod ile kategeorik doldurma yapılmasının sebebi en çok tekrar eden değerin eksik verilerin yerini alacak olmasıdır.
df["Names"].fillna(df["Names"].mode()[0], inplace=True)   ## mode()[0] tekrar edenler arasından birinciyi seçer 
```

```python
# Gelişmiş İmputasyonlar


# KNN İmputasyon: Komşuluk Algoritması ile doldurmak

"""


KNN algoritması, her eksik değeri, o noktaya en yakın olan k komşusunun değerlerine dayalı olarak tahmin eder.
Bu, verinin komşuluk ilişkilerini dikkate alarak eksik veriyi doldurur.
KNN Algoritması sadece sayısal verilerle çalışmaktadır.
    -> kategorik verileri label_encoder veya onehot yöntemleriyle sayısal değişkene çevirebilir
    -> sadece sayısal değişkenleri işleme alarakta bir uygulama yapabilirsin.
    
Örneğin, bir veri çerçevesinde bir kişinin yaşını tahmin etmek için aynı özelliklere sahip diğer kişilerin yaşları göz önüne alınabilir

"""



"""

Kategorik verileri dönüştürme işlemi yapmak isteyenler için:


from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Names'] = label_encoder.fit_transform(df['Names'])

"""


# kütüphane import
from sklearn.impute import KNNImputer

# sayısal değişkenleri seçmek
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# df_numeric = df.select_dtypes(exclude=['object'])

# KNN İmputasyonu Uygulamak
imputer = KNNImputer(n_neighbors=3)

# Modeli fit etmek
df_imputed = imputer.fit_transform(df_numeric)

# Sonuçları tekrar dataframe'e çevirmek
df_imputed = pd.DataFrame(df_imputed, columns=df_numeric.columns)
df_imputed.head()
```

```python
# Regresyon İmputasyonu : Eksik veriler, regresyon modeli kullanılarak tahmin edilebilir. 
# Bu yöntemde, eksik verinin yerine doldurulacak değer, diğer değişkenlere dayanarak tahmin edilir.

# Bu yaklaşım özellikle eksik verilerin belirli bir ilişkiyi takip ettiği ve diğer özelliklerden yararlanarak 
# tahmin edilebileceği durumlarda oldukça etkili olabilir.

# Örneğin, bir değişkenin (yaş) eksik olduğu durumda, diğer değişkenler (gelir, eğitim durumu vb.) kullanılarak yaş tahmin edilebilir.


import pandas as pd
from sklearn.linear_model import LinearRegression  # reg modellerinden -> basit regresyon import işlemi
from sklearn.model_selection import train_test_split  # model seçimi için egitim-test importu işlemi
from sklearn.impute import SimpleImputer # simple reg (basit regresyon )imputer import etme işlemi

# Örnek veri çerçevesi
data = {
    'Age': [25, 30, 35, 40, None, 50, None, 60, 70],
    'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000],
    'Experience': [2, 4, 6, 8, 10, 12, 14, 16, 18]
}

df = pd.DataFrame(data)

# Eksik veriye Bakmak
df.isnull().sum()

# 1. Adım: Eksik olan 'Age' sütununu tahmin etmek için regresyon modelini eğitmek

# Eksik veriyi dolduracağımız sütunu belirleyelim: 'Age'
df_with_missing = df[df['Age'].isnull()]  # Eksik verilerin olduğu satırların seçilmesi
df_no_missing = df.dropna(subset=['Age'])  # Eksik verisi olmayan satırların seçilmesi

# 2. Adım: 'Salary' ve 'Experience' sütunlarını bağımsız değişkenler olarak kullanacağız
X_train = df_no_missing[['Salary', 'Experience']]  # Bağımsız değişkenler
y_train = df_no_missing['Age']  # Bağımlı değişken ('Age')

# 3. Adım: Regresyon modelini oluşturup eğitelim
model = LinearRegression()
model.fit(X_train, y_train)

# 4. Adım: Eksik 'Age' değerlerini tahmin edelim
X_missing = df_with_missing[['Salary', 'Experience']]  # Eksik verileri olan satırların bağımsız değişkenleri
df_with_missing['Age'] = model.predict(X_missing)  # Tahmin edilen 'Age' değerini yerine koy

# 5. Adım: Sonuçları birleştirelim
df_imputed = pd.concat([df_no_missing, df_with_missing])

df_imputed


# Linear Regression: Basit lineer regresyon örneğiyle bu işlemi gerçekleştirdik, 
# ancak farklı regresyon modelleri (örneğin, Ridge, Lasso) ve daha karmaşık yöntemler de kullanılabilir.
```

```python
# Multiple Imputation (Çoklu İmputasyon):

# Bu yöntem, her eksik veri için birden fazla tahmin yapar ve sonra bu tahminlerin ortalamasını alır. 
# Bu sayede, belirsizliği daha iyi yansıtan ve güvenilir sonuçlar elde edilir.
# Multiple Imputation, eksik veriyi tahmin ederken farklı olasılıkları dikkate alır ve her tahmin için birden fazla değer üretir.

# Bu yaklaşım, özellikle eksik verilerin rastgele olmayan (MNAR - Missing Not at Random) durumlarda kullanışlıdır.

import pandas as pd
from fancyimpute import IterativeImputer  ## kütüphane import


data = {
    'Age': [25, 30, 35, 40, None, 50, None, 60, 70],
    'Salary': [50000, 60000, 70000, None, 90000, 100000, 110000, 120000, None],
    'Experience': [2, 4, 6, 8, 10, 12, 14, 16, 18]
}

df = pd.DataFrame(data)

# Eksik Verilere bakalım
df.isnull().sum()

# Çoklu imputasyon işlemi 

# iterative, belirli koşula göre tekrar edilmek anlamına gelir.
# max 10 defa tekrar edilecek iterasyon ile imputer yapılacak
imputer = IterativeImputer(max_iter=10, random_state=0)  


# Veriyi imputasyon ile doldur
df_imputed = imputer.fit_transform(df)

# Sonuçları bir DataFrame'e dönüştürelim
df_imputed = pd.DataFrame(df_imputed, columns=df.columns)

df_imputed
```

## Korelasyon Analizi ve Aykırı Değerler

Korelasyon : Korelasyon iki değişken arasındaki ilişkinin yönünü ve gücünü ifade eden bir istatistiksel ölçüdür.  

Korelasyon Özellikleri
 * Pozitif Korelasyon : Bir değişken arttığında diğeri de artıyorsa pozitif korelasyon (1)  
 * Negatif Korelasyon : Bir değişken arttığında diğeri azalıyorsa negatif korelasyon (-1)  
 * Sıfır Korelasyon : Değişkenler arasında anlamlı bir ilişki yoksa sıfır korelasyon vardır (0).  

Korelasyon Türleri
1. Pearson Korelasyonu 
 * En yaygın kullanılan korelasyon türüdür.
 * Değişkenlerin her ikisinin de sürekli ve normal dağılıma sahip olduğu durumlarda kullanılır.

2. Spearman Korelasyonu
 * Verilerin sıralanmasına dayalı bir korelasyon türüdür.
 * Özellikle verilerin sıralı olduğu ve normal dağılıma uymadığı durumlarda kullanılır.

 ### Korelasyon ve Nedensellik

Korelasyon, iki değişkenin birbiriyle ilişkili olduğunu gösterir, ancak birinin diğerine neden olduğunu kanıtlamaz.  
Bu nedenle, korelasyonun yüksek olması, nedensellik olduğunu varsaymak için yeterli değildir.  

Örnek:  
Yazın dondurma satışları ile boğulma vakalarının artışı arasında pozitif korelasyon vardır.  
Ancak dondurma satışlarının boğulmalara neden olduğunu söylemek doğru değildir.  
Her iki olayı da sıcak hava etkiler.


```python
import pandas as pd
import numpy as np

# Örnek veri seti
data = {
    'X': [10, 20, 30, 40, 50],
    'Y': [15, 25, 35, 45, 55]
}
df = pd.DataFrame(data)

# X  ve Y arasındaki Korelasyonu hesaplama
pearson_corr = df['X'].corr(df['Y'])  # Pearson korelasyonu
spearman_corr = df['X'].corr(df['Y'], method='spearman')  # Spearman korelasyonu

print("Pearson Korelasyonu:", pearson_corr)
print("Spearman Korelasyonu:", spearman_corr)
```

```python
# Örnek Uygulama

import pandas as pd

data = {
        'Yaş': [25, 30, 35, 40, 45], 
        'Gelir': [2000, 3000, 4000, 5000, 6000]
        }

df = pd.DataFrame(data)


pearson_corr = df['Yaş'].corr(df['Gelir'])
pearson_corr
```

 * Pearson Korelasyonu: Doğrusal ilişkiyi ölçer

Yaş artarken Gelir değeri de artış göstermektedir.

### Korelasyon vs Regresyon 

Korelasyon ve regresyon, iki değişken arasındaki ilişkiyi anlamaya yönelik istatistiksel araçlardır.   
Ancak, bu yöntemler farklı amaçlar için kullanılır ve sonuçlarının yorumlanması da farklıdır.  

Korelasyon, iki değişken arasındaki ilişkinin gücünü ve yönünü ölçer.  
Örnek : Gelir ve Harcama arasındaki ilişkiyi anlamak.  

Regresyon, Bir değişkenin diğerini nasıl etkilediğini modellemek ve tahmin etmek için kullanılır.  
Örnek : Gelir düzeyine göre harcamayı tahmin etmek.

```python
# Regresyon Örneği

# Yaş'a göre Gelir Tahmin Modeli 

import pandas as pd
from sklearn.linear_model import LinearRegression 

# Veri seti
data = {
    
    'Yaş': [25, 30, 35, 40, 45], 
    'Gelir': [2000, 3000, 4000, 5000, 6000]
    
       }

df = pd.DataFrame(data)


X = df[['Yaş']]  # Bağımsız değişkenimiz
y = df['Gelir']  # Bağımlı değişkenimiz

# Linear Regression (Doğrusal İlişki) Modeli Oluşturmak
# Yaş ve Gelir arasında doğrusal ilişki olduğu için Linear Regression kullanıldı.
model = LinearRegression()
model.fit(X, y)

# Regresyon katsayıları
print("Eğitim Katsayısı (b):", model.coef_[0])
print("Sabit Terim (a):", model.intercept_)

# Tahmin, yaş 50 olursa gelir kaç olur?
yas_tahmin = 50
gelir_tahmin = model.predict([[yas_tahmin]])

print(f"{yas_tahmin} yaşında tamin edilen gelir: {gelir_tahmin[0]}")
```

### 3.3 Outlier (Aykırı Değer) Tespiti ve Yöntemi
Aykırı değerler, veri setinde diğer gözlemlerden önemli ölçüde farklı olan verilerdir.  
Bu tür verilerin tespiti ve uygun yöntemlerle düzeltilmesi gerekir.  
Aykırı değerler, modelin performansını olumsuz etkileyebilir.

#### Aykırı Değer için Dikkat Edilmesi Gerekenler:

* Aykırı Değerin Tanımı: Aykırı değerler veri kümesinin özelliklerine bağlıdır.  
Bir veri kümesinde aykırı olan bir değer, başka bir veri kümesinde normal kabul edilebilir.

* Etkisi: Aykırı değerler, model parametrelerini, dağılımı, ve tahmin sonuçlarını ciddi şekilde etkileyebilir.

* Sebep Analizi: Aykırı değerlerin yanlış ölçüm, veri giriş hatası veya doğal fenomen kaynaklı olup olmadığını anlamak önemlidir.

* Yönetim Kararı: Aykırı değerlerin yönetimi, verinin türüne ve projedeki önceliklere göre değişir.  
Tamamen kaldırmak yerine, değiştirme veya yeniden ölçeklendirme yöntemleri tercih edilebilir.  

#### Aykırı Değer Tespit Yöntemleri

* İstatistiksel Yöntemler : Z-score (Standart Sapma Yöntemi), IQR (Çeyreklikler Arası Genişlik)
* Görselleştirme Yöntemleri : Boxplot, Histogram  
* Makine Öğrenmesi Yöntemler : Isolation Forest, DBSCAN 

```python
# Z-Skor Yöntemi (Standart Sapma Yöntemi)

# Z-Skor > 3  veya Z-Skor < -3 olan değerler aykırı değer olarak kabul edilmektedir.

# Z-Score = [(Value - Average of Value)] / STD

# Örnek veri ile uygulama
data = {"X": [10, 12, 14, 15, 16, 17, 18, 300]}
df = pd.DataFrame(data)

# Z-skoru hesaplama
# Z-Score = (X - X'in Ortalaması) / X'in Standart Sapması

df["Z-Score"] = (df["X"] - df["X"].mean()) / df["X"].std()
aykiri_degerler = df[df["Z-Score"].abs() > 3]
print(aykiri_degerler)


```

```python
# IQR (Interquartile Range – Çeyrekler Arası Genişlik)

# IQR, verilerin dağılımını inceleyerek aykırı değerleri tespit eder.

# IQR = Q3 - Q1 (Ücüncü Çeyrek - Birinci Çeyrek)

# Aykırı Değerler : [Q1 − 1.5 × IQR] ve [Q3 + 1.5 × IQR] sınırlarının dışındadır.
# Alt ve Üst sınır olarak belirlenir ve dışında kalanlar aykırı kabul edilirler.

# IQR Yöntemi Örnek Uygulama

data = {"Value": [10, 12, 14, 15, 16, 17, 18, 300]}
df = pd.DataFrame(data)

# Value'nin 1. ve 3. çeyrekliklerini atama işlemi
Q1 = df["Value"].quantile(0.25)
Q3 = df["Value"].quantile(0.75)

# IQR (Çeyrekliklerin arasındaki genişliğin hesaplanması)
IQR = Q3 - Q1

# alt ve üst sinir yani çeyrekliklerin dışında kalan kısımların belirlenmesi
alt_sinir = (Q1 - 1.5 * IQR)
ust_sinir = (Q3 + 1.5 * IQR)

# VALUE değeri alt sınırdan küçükse,
# VALUE değeri ust sınırdan büyükse aykiri değer kabul edilir ve aykiri değer değişkenine atanır.
aykiri_degerler = df[(df["Value"] < alt_sinir) | (df["Value"] > ust_sinir)]

aykiri_degerler.head()
```

```python
# Görselleştirme Yöntemleri

# Boxplot: Çeyrekler arasındaki dağılımı ve aykırı değerleri hızlıca görselleştirir.
# Histogram: Veri dağılımını göstererek uç değerlerin fark edilmesini sağlar.

sns.boxplot(x=df["Value"], data = df)
```

```python
# Makine Öğrenmesi ile Aykırı Değer Tespiti

# Isolation Forest : Aykırı değerleri tespit etmek için ağaç tabanlı bir algoritma kullanır.

# kütüphane import işlemi
from sklearn.ensemble import IsolationForest

# Isolation Forest ile Model
iso = IsolationForest(contamination=0.1)  # contamination, veri setinin %10'luk dilimini aykiri kabul ederek hareket edecek

# Modeli fit edeceğiz ve aykiri_degere atayacağız
df["Aykiri_Deger"] = iso.fit_predict(df[["Value"]])

# aykırı değeri -1 olarak işaretleyecek
df 
```

```python
# Alt ve Üst Sınırları Kullanarak Aykırı Değerleri Yönetmek
# (Clipping)

# Alt ve üst sınır belirlemiştik

# Value alt sınırdan küçükse, alt sınıra eşitleyecek
# Value üst sınırdan büyükse, üst sınıra eşitleyecek 
# Aksi takdirde olduğu gibi bırakacak ve aykırı değerleri kurtaracağız

df["Value"] = np.where(df["Value"] > ust_sinir, ust_sinir,
                       np.where(df["Value"] < alt_sinir, alt_sinir, df["Value"]))

df 
# aykırı değeri 22.88 olarak düzenlendi
```

### 3.4 Veri Ölçeklendirme ve Normalizasyon

Veri ölçeklendirme ve normalizasyon, veri ön işleme süreçlerinde sıkça kullanılan iki önemli tekniktir.  
Bu yöntemler, veriyi belirli bir ölçeğe getirerek daha etkili analiz ve modelleme yapılmasını sağlar.  
İkisi de farklı amaçlara hizmet eder ve farklı durumlarda kullanılır.

#### 3.4.1. Ölçeklendirme : 
Verilerin belirli bir aralığa ((0,1) veya (-1, 1)) veya ölçeye göre yeniden düzenlenmesidir.  
Bu yöntem genellikle verilerin farklı birimlerde veya büyük ölçekte değişiklik gösterdiği durumlarda kullanılır.
 
 * Min-Max Scaling : Veriyi [0,1] veya başka bir aralığa göre çeker. (Özellikle lojistik reg. ve SVM gibi mesafe tabanlı modellerde)  

 * Standartlaştırma : Verilerin ortalamasını 0 ve standart sapmasını 1 yapar. (özlk. normal dağ. varsayan algoritmalarda > PCA, K-Means)  

 * Robust Scaling : Aykırılıklara duyarsız olduğu için median ve IQR kullanılarak gerçekleştirilir. (Aykırılıkların fazla olduğu veri setleri)  


#### 3.4.2 Normalizasyon :   
Verilerin büyüklüğünü standart bir formata getirmek için kullanılır.  
 her veri noktasını diğer veri noktalarına göre ölçeklendirmeyi ifade eder.

 * L1 Normalizasyon (Manhattan Norm) : Her bir değeri, vektörün toplam değerine böler. (Özellikle parça-bütün ilişkilerinde kullanılır)

 * L2 Normalizasyon (Euclidean Norm) : Her bir değeri, öklid normuna böler (özellike mesafe tabanlı modellerde örn. K-NN kullanılır.)

#### Dikkat Edilmesi Gereken Noktalar

Ölçeklendirme vs Normalizasyon  
Ölçeklendirme, genellikle verilerin birimlerini veya aralığını değiştirmeyi içerirken, normalizasyon vektörün büyüklüğünü standardize eder.

Algoritma uyumluluğu:

 * Mesafe veya gradyan tabanlı algoritmalarda (SVM, k-NN, k-Means, PCA) Ölçeklendirme gereklidir.
 * Derin öğrenme modellerinde Normalizasyon sıklıkla tercih edilir.
 * Eğer veri setinde çıkıntılar varsa, Robust Scaling veya Standartlaştırma daha iyi çalışabilir


Veriyi ölçeklendirme ve normalizasyon, makine öğrenmesi modellerinin performansını artırabilir.  
Kullanılacak yöntemin seçimi, verinin özelliklerine ve kullanılacak algoritmaya bağlıdır.  
Her iki yöntem de, veriyle çalışan uzmanların mutlaka öğrenmesi ve uygulaması gereken temel ön işleme tekniklerindendir.  
 

```python
# Min - Max Scaler Uygulama
# Veriyi [0,1] veya başka bir aralığa göre çeker. (Özellikle lojistik reg. ve SVM gibi mesafe tabanlı modellerde)  


from sklearn.preprocessing import MinMaxScaler  #tüm veri ön işleme (veri hazırlığı) kütüphaneleri sklearn.preprocessing içerisinde yer alıyor

# ölçeklendirme yapılacak veri seti
data = np.array([[2, 4], [6, 8], [10, 12]])

# ölçeklendirmenin aralığı, scaler'e atanacak
scaler = MinMaxScaler(feature_range = (0, 1))   # aralık 0,1 arasında olacak

# scaler fit etme işlemi
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

```python
# Standartlaştırma  
# Verilerin ortalamasını 0 ve standart sapmasını 1 yapacak şekilde verileri düzenler. 
# (özlk. normal dağ. varsayan algoritmalarda > PCA, K-Means) 


from sklearn.preprocessing import StandardScaler

# model atama işlemi
scaler = StandardScaler() 

# model fit etme işlemi
standardized_data = scaler.fit_transform(data)

standardized_data
print("Veri setinin ortalaması : ",standardized_data.mean())
print("Veri setinin standart sapması : ",standardized_data.std())


```

```python
# L2 (Euclidean Norm) 
# Her bir değeri, öklid normuna böler (özellike mesafe tabanlı modellerde örn. K-NN kullanılır.)

from sklearn.preprocessing import normalize

normalized_data = normalize(data, norm='l2')

normalized_data

```

## 3.5 Özellik Mühendisliği (Feature Engineering)

Özellik Mühendisliği (Feature Engineering), bir veri bilimi projesinin temel adımlarından biridir ve modelin doğruluğunu artırmak için verilerden yeni özellikler oluşturmayı veya mevcut özellikleri dönüştürmeyi içerir.  

Bu süreç, ham veriden anlamlı ve yararlı bilgiler çıkarmayı amaçlar, böylece makine öğrenmesi modelleri daha doğru ve etkili sonuçlar verebilir.

Dikkat Edilmesi Gereken Diğer Noktalar:
* Özellik mühendisliği, veri setine özgü bir süreçtir; her problem için kullanılan özellikler farklı olabilir.
* İyi yapılmış özellik mühendisliği, modelin başarısını doğrudan etkiler.
* Özellik mühendisliğinde kullanılan teknikler, genellikle istatistiksel analiz ve alan bilgisi gerektirir.

Özellik mühendisliğini daha iyi anlamak için iki veri seti oluşturulup, bunlar üzerinde örnekler yapılacak.

```python
# Satış Verisi
sales_data = pd.DataFrame({
    'sale_id': [1, 2, 3, 4, 5],
    'customer_id': [101, 102, 103, 101, 104],
    'product': ['Laptop', 'Phone', 'Tablet', 'Phone', 'Laptop'],
    'price': [1500, 800, 600, 800, 1500],
    'quantity': [2, 1, 3, 1, 1],
    'date': pd.to_datetime(['2024-12-01', '2024-12-02', '2024-12-03', '2024-12-04', '2024-12-05'])
})

# Müşteri Verisi
customer_data = pd.DataFrame({
    'customer_id': [101, 102, 103, 104],
    'name': ['John', 'Sarah', 'Michael', 'Anna'],
    'age': [28, 35, 22, 40],
    'gender': ['Male', 'Female', 'Male', 'Female'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Boston']
})

# Veri Setlerini Merge Etme
df = pd.merge(sales_data, customer_data, on='customer_id', how='inner')

# Eğer ortak column olmasaydı
# df = pd.merge(sales_data, customer_data, left_on='sale_id', right_on='customer_id', how='inner')
# sol taraftan sales_customer_id ve sağ taraftan customer_id columnlarını alarak birleştir diyebilirdik

df
```

```python
# Toplam Satış Miktarını Hesaplama: 
# Her bir satışın toplam fiyatını (price * quantity) hesaplayarak, toplam satış miktarını bulalım.

df['total_sales'] = df['price'] * df['quantity']

df[['sale_id', 'total_sales']]
```

```python
#Yaşa Göre Toplam Satış: Müşterilerin yaşlarına göre toplam satışları hesaplayalım.
age_sales_sum = df.groupby('age')['total_sales'].sum().reset_index()

#Yaşa Göre Ortalama Satış: Müşterilerin yaşlarına göre ortalama satışları hesaplayalım.
age_sales_average = df.groupby('age')['total_sales'].mean().reset_index()

print("Yaşlara Göre Toplam Satış ", age_sales_sum)
print("Yaşlara Göre Ortalama Satış", age_sales_average)
```

```python
# Cinsiyet ve Ürün Bazında Satış Analizi: Cinsiyet ve ürün türüne göre toplam satışları inceleyelim.

# df.groupby([]) [] -> ilk parantez () groupby fonks. parantezidir 
# içerisindeki [] ise iki parametre alacağı içindir
# ikinci parantez her zaman [] olarak gelir
gender_product_sales = df.groupby(['gender', 'product'])['total_sales'].sum().reset_index()

gender_product_sales
```

```python
# Şehir Bazında Satış Performansı: Şehirlere göre toplam satışları inceleyelim.

city_sales = df.groupby("city")["total_sales"].sum().reset_index()

city_sales
```

```python
# Yaş grubu kategorilerini oluşturma

# pd.cut ile age değişk. 0-18, 18-35, 35-50, 50-100 aralıklarına bins ile ayırdık 
# ayırdıgımız kısımlara ise labels ile isimlendirme yaptık
df["AgeCategory"] = pd.cut(df["age"], bins=[0, 18, 35, 50, 100], labels=["Young", "Youth", "Adult", "Senior"])


df[['customer_id', 'age', 'AgeCategory']]
```

```python
# Zaman Özellikleri : tarih bilgisinden gün, hafta, ay bilgileri çıkartılabilir.
# Böylece her bir satışın hangi gün ve hangi ayda yapıldığını gösteren yeni sütunlar oluşturulabilir.

# date.day_name() fonks. ile Haftanın günlerini getireceğiz ve yeni oluşturduğumuz day_of_week değişk. atayacağız.  
df['day_of_week'] = df['date'].dt.day_name()

# date.month() fonks. ile ayları  getireceğiz ve yeni oluşturduğumuz month değişk. atayacağız.  
df['month'] = df['date'].dt.month

# date.year() fonks. ile yılları  getireceğiz ve yeni oluşturduğumuz year değişk. atayacağız.
df["year"] = df["date"].dt.year

# Tüm bu yeni değerlerden yeni bir table 
df[['sale_id', 'day_of_week', 'month', "year"]]
```

```python
# Label Encoding: Kategorik veriyi sayısal değerlere dönüştürme.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])

df
```

# Veri Ön İşleme
## Veri Ön İşlemeye Genel Bakış

Bu konuya başlamadan önce "Veri mi, Model mi?" sorusu konuyu özetler niteliktedir.

Biliyoruz ki kurulan makine öğrenmesi modelinin yüksek güvenilirlikle çalışması için doğru veriler gerekmektedir.  
Doğru veriler için de öncelikle Veri Ön İşleme gerekmektedir.  

* Makine Öğrenmesi Modeli'nin amacı genellenebilir yapılar ortaya koymaktır.  
* Belirli olaylar gözlemlendiğinde belirli sonuçlar vermektedir.    

(Harvard Business Review: Eğer veriniz kötü ise, makine öğrenmesi araçlarınız kullanışsız olacaktır.)  

## Veri Ön İşlemeye Genel Bakış

1) **Veri Temizleme (Data Cleaning / Cleasing)**

* Gürültülü Veri (bir erkeğin veri tabanında hamilelik durumunun "Evet" olarak girilmesi)
* Eksik Veri Analizi (N/A değerlerin girilmesidir.)
* Aykırı Gözlem Analizi (bir bebek biberonu fiyatının 1.000.000 TL olarak girilmesi)


2) **Veri Standardizasyonu**

* 0 - 1 Dönüşümü (Evet = 1 ve Hayır = 0)
* Z-skoruna dönüştürme (Standart sapma yöntemine benzemektedir. Veri setinin sağına ve soluna örn. +- 2.5 değerleri eklenerek bir aralık belirlenmesidir.)
* Logaritmik dönüşüm (Verilerin daha simetrik hale gelmesini sağlar)


3) **Veri İndirgeme**

* Gözlem sayısının azaltılması
* Değişken sayısının azaltılması  


4) **Değişken Dönüşümleri**

Değişken dönüşümü, veri kümesindeki değişkenlerin dağılımını düzenlemek, ölçek farklılıklarını azaltmak ve modelleme performansını artırmak için yapılır.

* Sürekli değişkenlerde dönüşümler
* Kategorik değişkenlerde dönüşümler 

### Aykırı Gözlem Analizi (Outlier Analysis)

Aykırılık Nedir?:    
Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere aykırı gözlem denir.    

Aykırı Değer:  
Aykırılığı ifade eden nümerik değere aykırı değer denir.  
  

Genellenebilirlik kaygısıyla oluşturulan kural setlerini ya da fonksiyonları yanıltır.  
Yanlılığa sebep olurlar.



Yukarıda bulunan 3 aykırı değer grafiğin yönünü negatiften -> pozitife çevirmiştir!  






### Aykırılık Tespiti Nasıl Yapılır?  

Veride genel eğilimin oldukça dışına çıkan gözlemlerdir.  

Peki veri setinin genel eğiliminin dışına çıkmayı nasıl tanımlarız?  

1) Sektör Bilgisi  
* Örneğin bir ev fiyat tahmin modelinde 1000 metrekarelik evleri modellemeye almamak. (Çünkü alışılmışın dışında, ender rastlanan bir durumdur.)  



2) Standart Sapma Yaklaşımı
* Bir değişkeninin ortalamasının üzerine aynı değişkenin standart sapması eklenerek bulunan değere standart sapma yaklaşımı denir.  
Ortaya çıkan bu değer "eşik değer" olarak belirlenir ve bu değerin altındaki - üstündeki değerler aykırı değer olarak tanımlanır.  

Eşik Değer = ORTALAMA + 1x Standart Sapma  
Eşik Değer = ORTALAMA + 2x Standart Sapma  
Eşik Değer = ORTALAMA + 3x Standart Sapma  

Standart sapma yaklaşımında kaç tane standart sapma ekleyeceğinizi belirlerken, genellikle hedeflediğiniz güven aralığına ve analizdeki amacınıza bağlı olarak karar verirsiniz.  
Örneğin:  

- 1 Standart Sapma: Verilerin yaklaşık %68'i ortalama etrafında bu aralıkta bulunur.  
- 2 Standart Sapma: Verilerin yaklaşık %95'i bu aralıkta bulunur.  
- 3 Standart Sapma: Verilerin yaklaşık %99.7'si bu aralıkta bulunur.    



Eğer verilerinizi belirli bir güven aralığında incelemek istiyorsanız, bu yüzdeleri göz önünde bulundurarak kaç standart sapma ekleyeceğinize karar verebilirsiniz.  
Eğer verilerinizi yüksek güvenilirlikle test etmek isterseniz, örneğin 3 standart sapma kullanırsanız verileriniz büyük kısmı silinebilir!  



3) Z-Skor Yaklaşımı  
* Standart sapma yöntemine benzer şekilde çalışır. Değişken standart normal dağılıma uyarlanır, yani standartlaştırılır.  
Sonrasında -örneğin- sağından ve solundan +- 2.5 değerine göre eşik değer konulur ve bu değerlerin altında - üstünde kalan değerlere aykırı değer denir.      
  
  

4) Boxplot (Interquartile range - IQR Yöntemi)
* En sık kullanılan yöntemlerden birisidir. Değişkenleri küçükten büyüğe sıralanır,  
Çeyrekliklerine (yüzdeliklerine) yani Q1 ve Q3 değerlerine karşılık gelen değerler üzerinden bir eşik değer hesaplanır. Bu eşik değere göre aykırı değer tanımı yapılır.

    
   

Seçim Yaparken Dikkat Edilmesi Gerekenler:

Verinin dağılımı ve bağlamı: Hangi yöntemin verinizin doğasına uygun olduğunu belirlemek için önemlidir.
Aykırı değerlerin analiz üzerindeki etkisi: Farklı teknikler, veri kümesindeki aykırı değerlerin ele alınma şeklini etkileyebilir.
Normal dağılım varsayımı: Z-skoru ve standart sapma yöntemleri genellikle bu varsayımı gerektirirken, IQR daha genel durumlar için uygundur.  
Z-skoru ve standart sapma yönteminde dağılımın normal olduğundan emin olunmalıdır.  

Veri Setinin Dağılımının İncelenmesi:  
* Histogram Grafiği  
* Shapiro - Wilk Testleri  
* Boxplot

# Aykırı Değer Yakalamak

```python
# Aykırı Gözlem Analizi

import seaborn as sns
import pandas as pd
import numpy as np

diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.select_dtypes(include=['float64', 'int64'])
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
# Öncelikle tek değişken üzerinde aykırı değer yakalama işlemi yapacağız (TABLE değişkeni)

df_table = df["table"]
df_table.head()
```

```python
sns.boxplot(x=df_table);
```

Grafiğin alt ve üst kısmındaki noktalara baktığımızda aykırı değerleri gözlemleyebiliyoruz.  
Eşik değer belirlenerek bu alt ve üst kısımdaki aykırılıkları yakalama işlemi yapacağız.

```python
Q1 = df_table.quantile(0.25)
Q3 = df_table.quantile(0.75)

IQR = Q3 - Q1
```

```python
alt_sinir = Q1 - (1.5 * IQR)
ust_sinir = Q3 + (1.5 * IQR)
```

```python
alt_sinir # 51.5
ust_sinir # 63.5 olarak gelmektedir.
```

```python
# Bu değişkende bulunan ve alt sınırdan küçük olan veya üst sınırdan büyük olan değerlerin filtrelenmesi

df_table < alt_sinir
df_table > ust_sinir

# TRUE dönenler aykırı değerdir
```

```python
# Alt sınır ve üst sınır değerlerinin birlikte incelenmesi
# TRUE dönenler aykırı değer.

((df_table < alt_sinir) | (df_table > ust_sinir))
```

```python
# alt ve üst sınırların birleşimi ve aykırı_tf atanması 

aykiri_tf = (df_table < alt_sinir) | (df_table > ust_sinir)

aykiri_degerler = df_table[aykiri_tf]

aykiri_degerler
```

Alt ve üst sınırlar dahil olmak üzere tam 605 aykırı değer tespit edildi.
